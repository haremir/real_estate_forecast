```python
# Visualizer objesi oluştur
visualizer = BostonVisualizer(df)

# Eksik veri görselleştirme
visualizer.plot_missing_data(save_path=r"C:\Users\emirh\Desktop\mart3\src\visualizations\missing_data.png")

# Eksik veri sayısal analiz
missing_stats = df.isnull().sum().sort_values(ascending=False)
missing_stats[missing_stats > 0]
```

```python
# Korelasyon matrisi
visualizer.plot_correlation_matrix(
    save_path=r"C:\Users\emirh\Desktop\mart3\src\visualizations\scorrelation_matrix.png",
    annot_kws={"size": 10}
)

# Hedef değişkenle en yüksek korelasyonlu 5 özellik
corr_with_target = df.corr(numeric_only=True)['MEDV'].abs().sort_values(ascending=False)
corr_with_target[1:6]  # MEDV'yi hariç tut
```

```python
# MEDV dağılımı + boxplot
visualizer.plot_distribution(
    column="MEDV",
    save_path=r"C:\Users\emirh\Desktop\mart3\src\visualizations\medv_distribution.png",
    kde=True
)

# Aykırı değer analizi
q1 = df['MEDV'].quantile(0.25)
q3 = df['MEDV'].quantile(0.75)
iqr = q3 - q1
print(f"Aykırı değer sınırları: < {q1 - 1.5*iqr:.2f} veya > {q3 + 1.5*iqr:.2f}")
```

```python
# Oda sayısı-Fiyat ilişkisi
visualizer.plot_scatter(
    x_col="RM",
    y_col="MEDV",
    save_path=r"C:\Users\emirh\Desktop\mart3\src\visualizations\rm_vs_medv.png",
    trendline=True
)

# LSTAT-Fiyat ilişkisi (interaktif)
visualizer.plot_interactive_scatter(
    x_col="LSTAT",
    y_col="MEDV",
    color_col="RAD",  # Erişilebilirlik indeksi
    output_path=r"C:\Users\emirh\Desktop\mart3\src\visualizations\interactive_plot.html"
)
```

```python
# Pairplot (en önemli 4 özellik)
important_features = ['RM', 'LSTAT', 'PTRATIO', 'MEDV']
visualizer.plot_pairplot(
    columns=important_features,
    save_path=r"C:\Users\emirh\Desktop\mart3\src\visualizations\pairplot.png",
    diag_kind="kde"
)
```
